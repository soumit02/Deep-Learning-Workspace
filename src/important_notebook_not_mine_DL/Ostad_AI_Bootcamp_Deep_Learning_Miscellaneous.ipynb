# ANN (Fully Connected Network) Parameter Calculation



<img src="https://i.ytimg.com/vi/sM2Mm6aT_HI/maxresdefault.jpg" width="500"/>

In a simple Artificial Neural Network (ANN), parameters come only from **Dense (Fully Connected) layers**.

Each connection has:
- One weight
- One bias (per neuron)

---

### 1. Formula for Parameters in a Dense Layer

$$
\text{Parameters} = (N_{in} \times N_{out}) + N_{out}
$$

Where:
- $N_{in}$ = number of input neurons
- $N_{out}$ = number of output neurons

---

### 2. Why this formula?

Each output neuron:
- connects to all input neurons → $N_{in}$ weights
- has 1 bias

So for $N_{out}$ neurons:
- total weights = $N_{in} \times N_{out}$
- total biases = $N_{out}$

---

### 3. Example ANN

Architecture:

$$
10 \rightarrow 5 \rightarrow 3
$$

---

#### Layer 1: 10 → 5

$$
(10 \times 5) + 5
= 50 + 5
= 55
$$

---

#### Layer 2: 5 → 3

$$
(5 \times 3) + 3
= 15 + 3
= 18
$$

---

#### Total Parameters

$$
55 + 18 = 73
$$

---

### 4. Key Intuition

- ANN parameters grow **very fast** with input size
- First layer usually has the most parameters
- Flattened image inputs make ANN very large (one reason CNNs are preferred)

---

### 5. Quick Shortcut Table

| Layer Type | Parameters |
|------------|------------|
| Dense      | $$ N_{in} \times N_{out} + N_{out} $$ |
| Activation |$$ 0 $$|
| Dropout    | $$ 0 $$ |


````python
import tensorflow as tf
from tensorflow.keras import layers, models

# ANN model
model = models.Sequential([

    # Input layer (flattened 32x32 image)
    layers.Input(shape=(1024,)),

    # Hidden layer 1: 1024 -> 32
    layers.Dense(32, activation='relu'),

    # Hidden layer 2: 32 -> 16
    layers.Dense(16, activation='relu'),

    # Output layer: 16 -> 10
    layers.Dense(10, activation='softmax')
])

# Model summary
model.summary()

# Verify total parameters
print("Total params:", model.count_params())
````

# CNN Output Shape Calculation


<img src="https://miro.medium.com/v2/resize:fit:1400/1*A8cX1eHqGI7dIuH3awcjlw.jpeg" width="500"/>

For CNNs, you often need to compute the **output shape** of each layer before calculating parameters.

---

### 1. Convolution Layer Output Shape

For an input:

$$
H_{in} \times W_{in} \times C_{in}
$$

The output height and width are:

$$
H_{out} = \left\lfloor \frac{H_{in} - K + 2P}{S} \right\rfloor + 1
$$

$$
W_{out} = \left\lfloor \frac{W_{in} - K + 2P}{S} \right\rfloor + 1
$$

where:

- $K$ = kernel size
- $S$ = stride
- $P$ = padding

Output channels:

$$
C_{out} = \text{number of filters}
$$

So output shape:

$$
\boxed{
H_{out} \times W_{out} \times C_{out}
}
$$


---

#### Example 1

Input:

$$
32 \times 32 \times 3
$$

Conv:

- Kernel = 3×3
- Stride = 1
- Padding = 0
- Filters = 16


Calculate height:

$$
H_{out}=\left\lfloor\frac{32-3+2(0)}{1}\right\rfloor+1
$$

$$
H_{out}=30
$$


Width:

$$
W_{out}=30
$$


Output:

$$
\boxed{30 \times 30 \times 16}
$$


---

#### Example 2: Same Padding

Input:

$$
32 \times 32 \times 3
$$

Conv:

- Kernel = 3×3
- Stride = 1
- Padding = 1
- Filters = 16


$$
H_{out}=
\left\lfloor
\frac{32-3+2(1)}{1}
\right\rfloor+1
$$


$$
H_{out}=32
$$


$$
W_{out}=32
$$


Output:

$$
\boxed{32 \times 32 \times 16}
$$


---

### 2. Pooling Layer Output Shape

Pooling uses the same formula:

$$
H_{out}=
\left\lfloor
\frac{H_{in}-K+2P}{S}
\right\rfloor+1
$$


$$
W_{out}=
\left\lfloor
\frac{W_{in}-K+2P}{S}
\right\rfloor+1
$$


Channels remain unchanged:

$$
C_{out}=C_{in}
$$


Example:

Input:

$$
32 \times 32 \times 16
$$


MaxPool:

- Kernel = 2×2
- Stride = 2


$$
H_{out}=
\frac{32-2}{2}+1
$$


$$
H_{out}=16
$$


Output:

$$
\boxed{16 \times 16 \times 16}
$$


---

### 3. Flatten Layer

Flatten converts feature maps into a vector.


Input:

$$
8 \times 8 \times 32
$$


Calculation:

$$
8 \times 8 \times 32 = 2048
$$


Output:

$$
\boxed{2048}
$$


---

### 4. Dense Layer

Dense output:

$$
N_{out}
$$


Example:

$$
2048 \rightarrow 128
$$


Output:

$$
\boxed{128}
$$


---

### Shortcut Rules

#### Valid Padding ($P=0$)

For stride = 1:


$$
H_{out}=H_{in}-K+1
$$


$$
W_{out}=W_{in}-K+1
$$


Example:

$$
32 \rightarrow 30 \rightarrow 28 \rightarrow 26
$$


---

#### Same Padding

For stride = 1:


$$
H_{out}=H_{in}
$$


$$
W_{out}=W_{in}
$$


Only channels change:


$$
32\times32\times3
\rightarrow
32\times32\times64
$$


---

#### Pooling (2×2, stride=2)

Height and width are reduced by half:


$$
32 \rightarrow 16 \rightarrow 8 \rightarrow 4
$$


Channels stay the same.

---

### Complete CNN Example

```

Input:      32×32×3

Conv(3×3,16,s=1,p=1)
→ 32×32×16

MaxPool(2×2,s=2)
→ 16×16×16

Conv(3×3,32,s=1,p=1)
→ 16×16×32

MaxPool(2×2,s=2)
→ 8×8×32

Flatten
→ 2048

Dense(128)
→ 128

Dense(10)
→ 10

```

# CNN Parameter Count Calculation

In a CNN, the trainable parameters are the **weights and biases** that the model learns during training.

Layers with parameters:

1. Convolution Layer
2. Dense (Fully Connected) Layer

Layers without parameters:

- MaxPooling
- ReLU / Activation
- Flatten


---

### 1. Convolution Layer Parameters

A convolution layer contains filters.

Each filter has:

- Kernel weights
- One bias value


Formula:

$$
\text{Parameters}
=
(K_h \times K_w \times C_{in} \times C_{out})
+
C_{out}
$$


where:

- $K_h$ = kernel height
- $K_w$ = kernel width
- $C_{in}$ = input channels
- $C_{out}$ = number of filters


---

#### Example 1

Input:

$$
32 \times 32 \times 3
$$


Conv Layer:

- Kernel = 3×3
- Filters = 16


Weights:

$$
3 \times 3 \times 3 \times 16
$$


$$
=432
$$


Bias:

$$
16
$$


Total parameters:

$$
432+16=448
$$


So:

$$
\boxed{448 \text{ parameters}}
$$


---

### 2. Dense Layer Parameters

A dense layer connects every input neuron to every output neuron.


Formula:


$$
\text{Parameters}
=
(N_{in}\times N_{out})+N_{out}
$$


where:

- $N_{in}$ = input neurons
- $N_{out}$ = output neurons


---

Example:

Input:

$$
2048
$$


Dense:

$$
2048 \rightarrow 128
$$


Weights:

$$
2048 \times 128
$$


$$
=262144
$$


Bias:

$$
128
$$


Total:

$$
262144+128
$$


$$
=262272
$$


---

### 3. Pooling Layer Parameters

Example:

```

MaxPool(2×2)

```


Pooling only selects values.

No weights are learned.


$$
\boxed{0 \text{ parameters}}
$$


---

### 4. Flatten Layer Parameters


Flatten only reshapes the feature map.


Example:

$$
8 \times 8 \times 32
$$


Output:


$$
8 \times 8 \times 32=2048
$$


Parameters:

$$
\boxed{0}
$$


---

### Complete CNN Parameter Calculation


Architecture:


```

Input:      32×32×3

Conv(3×3,16,s=1,p=1)
→ 32×32×16

MaxPool(2×2,s=2)
→ 16×16×16

Conv(3×3,32,s=1,p=1)
→ 16×16×32

MaxPool(2×2,s=2)
→ 8×8×32

Flatten
→ 2048

Dense(128)
→ 128

Dense(10)
→ 10

```


---

#### Conv Layer 1

C_{in}=3

(3\times3\times3\times16)+16

#### Conv Layer 2

$$


$$
=4640
$$


---

#### Flatten


Feature map:

$$
8\times8\times32
$$


Flatten size:


$$
8\times8\times32=2048
$$


Parameters:


$$
0
$$


---

#### Dense Layer


Input:

$$
2048
$$

$$
128
$$

(2048\times128)+128
$$


$$
=262272
$$


---

10

\text{Total}

### Quick Memory Table

| Layer | Parameter Formula |
|---|---|
| Convolution | $$ K_h \times K_w \times C_{in} \times C_{out}+C_{out} $$ |
| Dense | $$ N_{in}\times N_{out}+N_{out} $$ |
| Pooling | $$ 0 $$ |
| Flatten | $$ 0 $$ |
| ReLU | $$ 0 $$|




````python
import tensorflow as tf
from tensorflow.keras import layers, models


# Complete CNN Architecture

model = models.Sequential([

    # Input: 32x32x3
    layers.Input(shape=(32, 32, 3)),

    # Conv(3x3, 16 filters, stride=1, padding=valid)
    # Output: 32x32x16
    layers.Conv2D(
        filters=16,
        kernel_size=(3,3),
        strides=1,
        padding="valid",
        activation="relu"
    ),

    # Output: 16x16x16
    layers.MaxPooling2D(
        pool_size=(2,2),
        strides=2
    ),

    # Conv(3x3, 32 filters, stride=1, padding=same)
    # Output: 16x16x32
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        strides=1,
        padding="same",
        activation="relu"
    ),

    # Output: 8x8x32
    layers.MaxPooling2D(
        pool_size=(2,2),
        strides=2
    ),

    # Output: 2048
    layers.Flatten(),

    # Dense: 2048 -> 128
    layers.Dense(
        128,
        activation="relu"
    ),

    # Output: 10 classes
    layers.Dense(
        10,
        activation="softmax"
    )
])


# Model Summary
model.summary()
````

# Image Augmentation

![](https://www.ibm.com/adobe/dynamicmedia/deliver/dm-aid--eaee14ad-c604-487e-a2c0-a8aa7d118fc2/data-augmentation-image-augment.png?preferwebp=true)

Image augmentation helps mainly because it **increases the diversity of your training data without actually collecting new data**.

Here’s what that really means in practice:

When you train a model (like a CNN), it learns patterns from the examples you show it. But real-world images vary a lot—objects can be shifted, rotated, brighter, darker, slightly zoomed, or partially occluded. If your training data is limited and “clean,” the model can easily **overfit**—it memorizes training images instead of learning general patterns.

Image augmentation fixes that by creating **modified versions of existing images**, such as:

* rotation (small angles)
* flipping (horizontal/vertical)
* zooming in/out
* shifting left/right/up/down
* changing brightness/contrast
* adding small noise

### Why this helps:

**1. Reduces overfitting**
Instead of seeing the same exact image repeatedly, the model sees many variations. This forces it to learn robust features instead of memorizing pixels.

**2. Improves generalization**
The model becomes better at handling real-world variations it didn’t explicitly see in training.

**3. Acts like “free data”**
You simulate a larger dataset without collecting or labeling new images.

**4. Builds invariance**
The model learns that a “cat is still a cat” even if:

* it’s flipped
* slightly rotated
* darker or brighter
* shifted in the frame

**Simple intuition:**

- Without augmentation → model learns: “this exact pattern = cat”

- With augmentation → model learns: “shape, texture, and structure = cat, regardless of small changes”

## Early Stopping

![](https://towardsdatascience.com/wp-content/uploads/2021/06/1W368qtCQlCYmt6YlwIYnJg.png)


Early stopping is a technique used during model training to **stop training when the model stops improving on validation data**.

Instead of training for a fixed number of epochs, you let the model train, but you watch validation performance (like validation loss or accuracy).

### Why it is useful:

* Prevents **overfitting** (model starts memorizing training data after a point)
* Saves **time and computation**
* Keeps the **best version of the model automatically**

### Simple idea:

During training:

* Training loss keeps going down
* Validation loss goes down, then may start going up

Early stopping says:

> “Stop training when validation loss doesn’t improve for a few epochs.”

That “few epochs” is called **patience**.

---

### Example intuition:

| Epoch | Train Loss | Val Loss                    |
| ----- | ---------- | --------------------------- |
| 1     | 0.9        | 0.8                         |
| 2     | 0.6        | 0.5                         |
| 3     | 0.4        | 0.4                         |
| 4     | 0.3        | 0.42 ← starts getting worse |
| 5     | 0.25       | 0.45                        |

Model should stop around epoch 3 or 4.



### TensorFlow code (Early Stopping)

```python
import tensorflow as tf

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',      # what to watch
    patience=3,              # wait 3 epochs before stopping
    restore_best_weights=True  # go back to best model automatically
)

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
```
